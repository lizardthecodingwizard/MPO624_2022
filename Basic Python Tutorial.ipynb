Now, let's make a very basic plot. Import the module matplotlib.

```python
import matplotlib.pyplot as plt
```

Define the x and y values as lists

```python
x = [1, 2, 3, 4, 5]
y = [5, 2, 4, 1, 3]
```

Now, we'll plot the points, label the axes, make a title, and show the plot!

```python
plt.plot(x,y)

plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.title('My first graph!')

plt.show()
```
